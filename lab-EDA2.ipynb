```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency, pearsonr
from scipy import stats



%matplotlib inline
pd.set_option('display.max_columns', None)

df = pd.read_csv("amz_uk_price_prediction_dataset.csv")
```

```python
df.head()
```

```python
cat_isbest = df[['category', 'isBestSeller']]
cat_isbest
```

```python
cat_isbest.value_counts()
```

```python
# Crear la tabla de contingencia
crosstab_result = pd.crosstab(df["category"], df["isBestSeller"], margins = False)
crosstab_result
```

```python
crosstab_result["Proportion"] = crosstab_result[True] / (crosstab_result[True] + crosstab_result[False])
```

```python
# Agrupa por 'category'
# Calcula el promedio de 'isBestSeller' en cada grupo
# Ordenar de mayor a menor proporción
proportion_best_sellers = (df.groupby('category')['isBestSeller'].mean().sort_values(ascending=False))
```

```python
proportion_best_sellers
```

```python
crosstab_sorted = crosstab_result .sort_values("Proportion", ascending=False)
crosstab_sorted
```

```python
# Calcular la prueba de Chi-cuadrado en la tabla de contingencia
chi2, p, dof, expected = chi2_contingency(crosstab_result)
```

```python
chi2, p, dof, expected 
```

```python
n = crosstab_result.sum().sum()  # Número total de observaciones
min_dim = min(crosstab_result.shape) - 1  # Menor dimensión entre las filas y las columnas
cramers_v = np.sqrt(chi2 / (n * min_dim))
cramers_v
```

```python
crosstab_result.plot(kind="bar", stacked=True, figsize=(14, 7), color=["skyblue", "lightcoral"])
```

```python
plt.title("Distribución de Best-Sellers por Categoría de Producto")
plt.xlabel("Categoría")
plt.ylabel("Número de Productos")
plt.xticks(rotation=90)  # Rotar las etiquetas de categoría
plt.show()
```

```python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python
df_no_outliers = df[(df["price"] >= lower_limit) & (df["price"] <= upper_limit)]
df_no_outliers
```

```python
top_20_categories = df_no_outliers['category'].value_counts().head(20).index
df_top_20 = df_no_outliers[df_no_outliers['category'].isin(top_20_categories)]

sns.violinplot(x='category', y='price', data=df_top_20)
plt.xticks(rotation=90)
plt.title('Distribución de Precios por Categoría')
plt.show()
```

```python
top_10 = df_no_outliers['category'].value_counts().head(10).index
top_10 = df_no_outliers[df_no_outliers['category'].isin(top_10)]
plt.figure(figsize=(14, 7))
sns.barplot(x="category", y="price", data=top_10, palette="muted")
plt.title("Distribución de Precios por Categoría de Producto")
plt.xticks(rotation=90)  # Rotar las etiquetas del eje X
plt.show()
```

```python
median_prices = df.groupby("category")["price"].median().sort_values(ascending=False)
median_prices
```

```python
# Calcular precios promedio por categoría
average_prices = df.groupby('category')['price'].mean().sort_values(ascending=False).head(10)

# Crear gráfico de barras
average_prices.plot(kind='bar', color ="green")
plt.title('Precio Promedio por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Precio Promedio')
plt.show()
```

```python
sns.boxplot(x='category', y='stars', data=df_top_20)
plt.xticks(rotation=90)
plt.title('Distribución de Calificaciones por Categoría')
plt.show()
```

```python
df_top_10_ratings = df[df["category"].isin(top_10)]

plt.figure(figsize=(14, 7))
sns.boxplot(x="category", y="stars", data=df_top_10_ratings, palette="Set3")
plt.title("Distribución de Calificaciones por Categoría de Producto (Top 10)")
plt.xticks(rotation=90)
plt.show()
```

```python
median_ratings = df.groupby("category")["stars"].median().sort_values(ascending=False)
median_ratings
```

```python
correlation, _ = pearsonr(df_no_outliers['price'], df_no_outliers['stars'])
correlation, _
```
